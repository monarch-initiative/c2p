```python
import os
import pandas as pd
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
```

```python
directory_path = '../../src/oncoexporter/ncit_mapping_files/cda_to_ncit_tissue_wise_mappings'

csv_files_dict = {}

for filename in os.listdir(directory_path):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory_path, filename)
        csv_files_dict[filename.replace('cda_to_ncit_map_', '').replace('.csv', '')] = pd.read_csv(file_path)
```

```python
def remove_rows_save_csv(df, iloc_numbers, file_path):
    df_dropped = df.drop(df.index[iloc_numbers])
    df_dropped.reset_index(drop=True, inplace=True)
    df_dropped.to_csv(file_path, index=False)

# remove_rows_save_csv(df, [0, 2, 5], 'path_to_save/new_dataframe.csv')
```
